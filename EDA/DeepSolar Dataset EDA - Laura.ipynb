```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data into pandas dataframe
deepsolar = pd.read_csv('../deepsolar_tract.csv', encoding='ISO-8859-1', index_col=0)
```

```python
deepsolar.head()
```

```python
# Calculate, sort, print proportion of missing data for each variable
missing_data = deepsolar.isnull().mean(axis=0).sort_values(ascending=False)
with pd.option_context('display.max_rows', None):
    print(missing_data)
```

```python
y = deepsolar.filter(['number_of_solar_system_per_household'], axis=1)
x = deepsolar.filter(['daily_solar_radiation'], axis=1)
plt.scatter(x,y)
plt.xlabel('daily_solar_radiation')
plt.ylabel("Solar System Density")
plt.show()
```
