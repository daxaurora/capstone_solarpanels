```python
num_rows=len(numeric_fields)/4
fig,ax=plt.subplots(int(num_rows),4,figsize=(16,150))
#fig.width(8)
item_index=0
for i in range(0,int(num_rows)):
    for j in range(0,4):
        data=relevant_fields[numeric_fields[item_index]].replace(np.inf,np.nan).dropna().values
        #print(numeric_fields[item_index],':')
        #print(data)
        ax[i,j].hist(data,bins=20)
        ax[i,j].set_title(numeric_fields[item_index])
        item_index+=1
plt.show()

```
